```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python

n = 19

qubo = (n)**2 
n_max = n

print(qubo)
x = np.arange(4, n_max)


plt.figure(figsize=(4 , 3), dpi = 200)

plt.grid(True, linestyle='--', alpha = 0.6)
plt.plot(
    x, (x)**2 , label = "conventional qubo", color = plt.cm.BuPu(0.35))
plt.plot(
    x, (x)*np.ceil(np.log2((x))) + (x)*( np.ceil(np.log2((x)))*(1/2)*(np.ceil(np.log2((x))) - 1)  ), label = "be qubo",
    color = plt.cm.Reds(0.45)
)
# plt.plot(
#     x, (x)*np.ceil(np.log2((x))) + (x)*( np.ceil(np.log2((x)))*(np.ceil(np.log2((x))))  ), label = "new qubo"
# )
plt.xlim(4,n_max-1)
plt.ylabel("# of variables ")
plt.xlabel("# of cities")
plt.hlines(y=177, xmin = 2, xmax = n_max , color="k", ls = "-.", label="DWave capability")
# plt.legend(loc = 4)
plt.show()
```
